```python
import pandas as pd
from fuzzywuzzy import process, fuzz

# Load the Excel files
flipkart_df = pd.read_excel(r"E:\DHP\Group_Project\amazon\Final_Amazon_Mobile.xlsx")
amazon_df = pd.read_excel(r"E:\DHP\Group_Project\Flipkart\Final_Flip_Mobile.xlsx")

# Check column names
print("Flipkart columns:", list(flipkart_df.columns))
print("Amazon columns:", list(amazon_df.columns))

# Match names using fuzzy matching and also store scores
def get_best_amazon_match(fk_name, amazon_names):
    match = process.extractOne(fk_name, amazon_names, scorer=fuzz.token_sort_ratio)
    if match:
        return pd.Series([match[0], match[1]])  # matched name, score
    return pd.Series([None, None])

# Apply the matching
flipkart_df[['matched_name', 'match_score']] = flipkart_df['Name'].apply(
    lambda x: get_best_amazon_match(x, amazon_df['Name'])
)

# Filter out matches below threshold 45
matched_df = flipkart_df[flipkart_df['match_score'] >= 90]

# Merge based on matched names
merged_df = pd.merge(
    matched_df,
    amazon_df,
    left_on='matched_name',
    right_on='Name',
    suffixes=('_flipkart', '_amazon')
)

# Build final dataframe
final_df = pd.DataFrame({
    'name': merged_df['Name_flipkart'],
    'price in amazon': merged_df['Price_amazon'],
    'price in flipkart': merged_df['Price_flipkart'],
    'rating in amazon': merged_df['Ratings_amazon'],
    'rating in flipkart': merged_df['Rating_flipkart'],
    'url in amazon': merged_df['Product URL'],
    'url in flipkart': merged_df['Product_URL'],
    'match_score': merged_df['match_score']  
})

# Save to Excel
final_df.to_excel(r"E:\DHP\Group_Project\Cleaned_file.xlsx", index=False)
print("✅ File saved successfully with", len(final_df), "matched products.")
```

```python
import pandas as pd
from fuzzywuzzy import process, fuzz

# Load the Excel files
amazon_df = pd.read_excel(r"E:\DHP\Group_Project\amazon\headphones.xlsx")
flipkart_df = pd.read_excel(r"E:\DHP\Group_Project\Flipkart\flip_headphones.xlsx")

# Function to get best match from Amazon for each Flipkart product name
def get_best_amazon_match(flip_name, amazon_names):
    match = process.extractOne(flip_name, amazon_names, scorer=fuzz.token_sort_ratio)
    return pd.Series([match[0], match[1]]) if match else pd.Series([None, None])

# Fuzzy match Flipkart names against Amazon names
flipkart_df[['matched_name', 'match_score']] = flipkart_df['Name'].apply(
    lambda name: get_best_amazon_match(name, amazon_df['Name'])
)

# Filter matches with score >= 90
matched_df = flipkart_df[flipkart_df['match_score'] >= 80]

# Merge the matched Flipkart and Amazon dataframes
merged_df = pd.merge(
    matched_df,
    amazon_df,
    left_on='matched_name',
    right_on='Name',
    suffixes=('_flipkart', '_amazon')
)

# Create the final formatted dataframe
final_df = pd.DataFrame({
    'Name': merged_df['Name_flipkart'],
    'Price in Amazon': merged_df['Price_amazon'],
    'Price in Flipkart': merged_df['Price_flipkart'],
    'Rating in Amazon': merged_df['Rating_amazon'],
    'Rating in Flipkart': merged_df['Rating_flipkart'],
    'URL in Amazon': merged_df['URL_amazon'],
    'URL in Flipkart': merged_df['URL_flipkart'],
    'Match Score': merged_df['match_score']
})

# Save the final comparison Excel file
final_df.to_excel(r"E:\DHP\Group_Project\Cleaned_H.xlsx", index=False)
print("✅ Final file saved with", len(final_df), "closely matched products.")
```
